```python
import win32com.client

def extract_text_from_shapes(doc_path):
    word = win32com.client.Dispatch("Word.Application")
    doc = word.Documents.Open(doc_path)

    text_boxes = []
    for shape in doc.Shapes:
        if shape.TextFrame.HasText:
            text = shape.TextFrame.TextRange.Text
            text_boxes.append(text)

    doc.Close(False)
    word.Quit()
    return text_boxes
```

```python
import win32com.client
import os

word = win32com.client.Dispatch("Word.Application")
word.Visible = False

# קבלת נתיב הקובץ
current_directory = os.getcwd()

# בניית הנתיב המלא
doc_path = os.path.join(current_directory, "7002.docx")

# הדפסת הנתיב כדי לוודא שהוא נכון
print("נתיב הקובץ:", doc_path)

# פתח את הקובץ
try:
    doc = word.Documents.Open(doc_path)
except Exception as e:
    print(f"שגיאה בפתיחת הקובץ: {e}")

# רשימה לאחסון הטקסטים
texts = []
for shape in doc.Shapes:
    if shape.TextFrame.HasText:
        texts.append(shape.TextFrame.TextRange.Text)

# סגירת הקובץ
doc.Close(False)
word.Quit()

# הדפסת התכנים
for i, text in enumerate(texts, 1):
    print(f"תיבת טקסט {i}: {text}")
```

```python
import win32com.client

# הנתיב לקובץ המקורי והחדש
input_path = r"C:\Users\User\Desktop\מסמכים לפרויקט\ממופה\7002.docx"
output_path = r"C:\Users\User\Desktop\מסמכים לפרויקט\ממופה\without_textboxes.docx"

# פתיחת Word
word = win32com.client.Dispatch("Word.Application")
word.Visible = False  # כדי ש-Word לא ייפתח על המסך

doc = word.Documents.Open(input_path)

# נאסוף את כל הטקסט מהצורות
text_from_shapes = []

for shape in doc.Shapes:
    if shape.TextFrame.HasText:
        text = shape.TextFrame.TextRange.Text.strip()
        if text:
            text_from_shapes.append(text)
        shape.Delete()  # להסיר את תיבת הטקסט עצמה

# הוספת הטקסט כתוכן רגיל למסמך בסוף
for text in text_from_shapes:
    doc.Content.InsertAfter("\n" + text)

# שמירה וסגירה
doc.SaveAs(output_path)
doc.Close()
word.Quit()

print("בוצע בהצלחה!")
```
